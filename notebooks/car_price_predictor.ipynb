# Car Price Predictor

## Introduction
This project aims to predict the selling price of used cars based on various features such as the car's name, company, year of purchase, kilometers driven, and fuel type. We utilize a **LinearRegressionModel** to establish a relationship between these features and the price.

## Problem Statement
The dataset contains mixed types of data (numerical and categorical) with some quality issues. The goal is to clean this data, engineer useful features, and train a regression model to predict `Price`.

## Dataset Overview
The dataset contains the following columns:
- **name**: Car model name
- **company**: Manufacturer
- **year**: Purchase year
- **Price**: Selling price (Target)
- **kms_driven**: Distance driven
- **fuel_type**: Fuel used (Petrol, Diesel, LPG)

## Methodology
1. **Data Loading**: Load the raw CSV.
2. **Exploratory Data Analysis**: Understand data distribution and issues.
3. **Cleaning**: Handle missing values, incorrect types, and format strings.
4. **Feature Engineering**: Extract relevant information from columns.
5. **Model Training**: Train a `LinearRegressionModel`.
6. **Evaluation**: Measure performance.

## Exploratory Data Analysis & Cleaning

```python
# Create a backup of the original data before cleaning
backup =car.copy() 
```

```python
# Filter out rows where 'year' is not numeric to ensure data integrity
car= car[car["year"].str.isnumeric()]
```

```python
car.shape
```

```python
# Convert 'year' to integer for numerical operations
car["year"] = car["year"].astype(int)
```

```python
car.info()
```

```python
# Remove rows with invalid price values ('Ask For Price')
car=car[car["Price"] != "Ask For Price"]
```

```python
car.shape
```

```python
# Clean 'Price' column: remove commas and convert to integer
car["Price"] = car["Price"].str.replace(",", "").astype(int)
```

```python
car.info()
```

```python
# Extract numerical part from 'kms_driven' and remove commas
car["kms_driven"] = car["kms_driven"].str.split(' ').str.get(0).str.replace(',', '')
```

```python
car.sample(5)
```

```python
car=car[car["kms_driven"].str.isnumeric()]
```

```python
car["kms_driven"]=car["kms_driven"].astype(int)
```

```python
car.info()
```

```python
# Remove rows with missing 'fuel_type' values
car=car[~car["fuel_type"].isna()]
```

```python
car["fuel_type"].unique()
```

```python
car.info()
```

```python
# Standardize 'name' by keeping only the first three words (Brand + Model + Submodel)
car["name"] = car["name"].str.split(" ").str.slice(0,3).str.join(" ")
```

```python
car.reset_index(drop=True, inplace=True)
```

## Feature Engineering & Preprocessing

```python
X = car.drop(columns='Price')
y = car['Price'] 
```

## Model Training (LinearRegressionModel)

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets to evaluate the LinearRegressionModel
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=433)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import  r2_score
from sklearn.preprocessing import  OneHotEncoder
# Create a column transformer to apply OneHotEncoding to categorical features leaving others as is
from sklearn.compose import make_column_transformer
# Create a pipeline combining preprocessing and the LinearRegressionModel
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
# Create a column transformer to apply OneHotEncoding to categorical features leaving others as is
colum_trans= make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
# Initialize the LinearRegressionModel
lr= LinearRegression()
```

```python
# Create a pipeline combining preprocessing and the LinearRegressionModel
pipe=make_pipeline(colum_trans,lr)
```

```python
# Train the LinearRegressionModel on the training data
pipe.fit(X_train , y_train)
```

```python
y_pred=pipe.predict(X_test)
```

## Evaluation

```python
r2_score(y_test,y_pred)
```

```python
import pickle

# Serialize the trained LinearRegressionModel for later use
pickle.dump(pipe,open("../model/model.pkl","wb"))
```
